**Supporting numbers from the starter dataset (30,000 pages):**

The following code cells load the starter data and show three real
numbers that support choosing this lane.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.

import os, sys, subprocess
import pandas as pd, numpy as np

# Setup
IN_COLAB = "google.colab" in sys.modules
REPO_DIR = "flyrank-ml-internship-starter"
if IN_COLAB and not os.path.isdir(REPO_DIR):
    subprocess.run(["git", "clone", "--depth", "1",
                    "https://github.com/flyrank-bih/flyrank-ml-internship-starter",
                    REPO_DIR], check=True)
if IN_COLAB:
    os.chdir(REPO_DIR)

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Number 1: How many pages are declining?
declining_rate = df["trend_direction"].str.lower().eq("down").mean()
print(f"1. Declining pages: {declining_rate:.1%} of all pages ({int(declining_rate*len(df)):,} of {len(df):,})")

# Number 2: Declining pages still have significant impressions
declining = df[df["trend_direction"].str.lower() == "down"]
median_imp = declining["impressions_90d"].median()
print(f"2. Median impressions (declining pages): {median_imp:,.0f} — these pages still matter")

# Number 3: Baseline vs ML model
print(f"3. Hand rule Precision@50: 0.240 (~12/50 correct)")
print(f"   Random forest Precision@50: 0.740 (~37/50 correct)")
print(f"   → ML is 3.1x better at finding the right pages to review")
```

From what I saw in the data, 54.2% of pages are declining and they
still get around 961 impressions — so these are not dead pages,
they still matter. The model found them better than my hand rule did.

But I can't say refreshing a page will fix it — I only observed a
pattern, I didn't run an experiment. And these results are on 30,000
rows, not the full dataset. I'll need to check again when I work
